```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import correlate
from scipy.fftpack import fft, fftfreq
```

Load data. Blank: nothing in the sample beam. Tape: bit of sellotape stuck to a microscope slide.

```python
# time (s), chA (V), chB (V)
blank = np.loadtxt('data/blank.txt', delimiter=',')
tape = np.loadtxt('data/tape.txt', delimiter=',')
```

Plots of the data as signal vs time.

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5), sharex=True, sharey=True)

for ax, data, title in zip((ax1, ax2), (blank, tape), ('Blank', 'Tape')):
    ax.plot(data[:,0],data[:,1], label='Reference')
    ax.plot(data[:,0],data[:,2], label='Sample')
    ax.set_xlabel('Time (s)')
    ax.set_ylabel('Photodetector signal (V)')
    ax.set_title(title)
    ax.set_xlim(right=data[-1,0])
    ax.legend(loc=8)
```

# Time delay

Calculate the time delay between the reference and sample beams for the blank and tape data.

```python
def time_delay(t, sigA, sigB):
    corr = correlate(sigA, sigB)
    t = np.concatenate((t[::-1]*-1,t[1::]))
    return (t, corr, t[corr.argmax()])

(blankt, blankcorr, bts) = time_delay(blank[:,0], blank[:,1], blank[:,2])
(tapet, tapecorr, tts) = time_delay(tape[:,0], tape[:,1],tape[:,2])

print('Blank time delay: %f s' % bts)
print('Tape time delay: %f s' % tts)
```

```python
ax = plt.axes()
ax.plot(blankt, blankcorr, label='blank')
ax.plot(tapet, tapecorr, label='tape')
ax.legend()
ax.set_ylabel('Cross-correlation (Vs)')
ax.set_xlabel('Time delay (s)')
```

Check that this time delay makes sense by shifting the sample beam by the calculated time delay. They should line up.

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5), sharex=True, sharey=True)

for ax, data, time_shift, title in zip((ax1, ax2), (blank, tape), (bts, tts),
                                       ('Blank + time shift', 'Tape + time shift')):
    ax.plot(data[:,0],data[:,1], label='Reference')
    ax.plot(data[:,0]+time_shift,data[:,2], label='Sample')
    ax.set_xlabel('Time (s)')
    ax.set_ylabel('Photodetector signal (V)')
    ax.set_title(title)
    ax.set_xlim(right=data[-1,0])
    ax.legend(loc=8)
```

They do!

# FFT
This will give the frequency of the rotation of the motor. It should be the same for all of the signals.

```python
def dofft(t, s):
    f = fftfreq(s.size, t[1]-t[0])
    F = fft(s)
    
    #http://gribblelab.org/scicomp/09_Signals_sampling_filtering.html
    ipos = f>0 
    f = f[ipos] # only look at positive freqs
    
    mags = abs(F[ipos]) # magnitude spectrum
    return (f, mags)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5), sharex=True, sharey=True)

frequency = [] # put most intense freq into this; they should all be the same

for ax, data, title in zip((ax1, ax2), (blank, tape), ('Blank', 'Tape')):
    print title
    
    (freq, fftA) = dofft(data[:,0],data[:,1])
    ax.plot(freq, fftA, label='Reference')
    maxfreqA = freq[fftA.argmax()]
    print 'A: %f Hz' % maxfreqA
    
    frequency.append(maxfreqA)
      
    (freq, fftB) = dofft(data[:,0],data[:,2])
    ax.plot(freq, fftB, label='Sample')
    maxfreqB = freq[fftB.argmax()]
    print 'B: %f Hz' % maxfreqB
    
    frequency.append(maxfreqB)
    
    ax.set_xlim(-100,2500)
    ax.set_title(title)
    ax.set_xlabel('Frequency (Hz)')
    ax.set_ylabel('Intensity')
    ax.legend()
```

Confirm that the frequencies are identical.

```python
np.all(frequency == frequency[0])
```

# Convert time delay to phase difference
First calculate the phase difference between the sample and reference beams for both the blank and the tape. Then calculate the difference in phase difference to get the rotation of the plane of polarized light by the sample. This is the answer.

```python
def rotation(time_delay,frequency):
    return time_delay * frequency * 360

blank_rotation = rotation(bts, frequency[0])
print 'Phase difference between reference and sample beams for blank: %.2f degrees' % blank_rotation

tape_rotation = rotation(tts, frequency[0])
print 'Phase difference between reference and sample beams for tape: %.2f degrees' % tape_rotation

print 'Difference between differences: %.2f degrees' % (blank_rotation - tape_rotation)
```
